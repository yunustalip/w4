```python
#Normalizasyon sonucunda orjinal deÄŸerlerinin deÄŸiÅŸeceÄŸi iÃ§in kolaylÄ±k olmasÄ± aÃ§Ä±sÄ±ndan orjinal CDRGLOB deÄŸerlerini bir kenarda tutuyoruz.
data["orjCDRGLOB"]=data["CDRGLOB"]

from sklearn import preprocessing #Normalizasyon uygulanarak veriler 0-1 aralÄ±ÄŸÄ±na sÄ±kÄ±ÅŸtÄ±rÄ±ldÄ±
data[['INDEPEND', 'CDRGLOB', 'APA', 'NACCMMSE', 'HIPPOVOL', 'LHIPPO', 'LENTM', 'RENTM']] = preprocessing.minmax_scale(data[['INDEPEND', 'CDRGLOB', 'APA', 'NACCMMSE', 'HIPPOVOL', 'LHIPPO', 'LENTM', 'RENTM']])
```

```python
# CDRGLOB ile korelasyonu en yÃ¼ksek olan bileÅŸenler arasÄ±ndan sadece aÅŸaÄŸÄ±daki 3 bileÅŸeni seÃ§menin doÄŸruluk oranÄ±nÄ±nÄ± maximize ettiÄŸi tespit edilmiÅŸtir.
data["MyFormul"] = data["INDEPEND"]*(0.693) + (data["NACCMMSE"])*(-0.7466) + data["RENTM"]*(-0.478) 
```
